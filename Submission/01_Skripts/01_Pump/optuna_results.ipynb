# Oxygen Pump

```python
import pandas as pd
import random
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("data/optimize_ox.csv")
df = df[df['Value'] != 0]

# Sort DataFrame by 'Value' column
df_sorted = df.sort_values(by='Value', ascending=False)

# Select top 10 highest values
df_top = df_sorted.head(10)

df_top
```

```python
# Define attributes and ranges
attributes = ['Param R', 'Param n', 'Param phi', 'Param psi']
ranges = {
    'Param R': (0.66, 0.678),
    'Param n': (35000, 39001),
    'Param phi': (0.14, 0.16),
    'Param psi': (0.68, 0.701),
}

# Create subplots
fig, axs = plt.subplots(len(attributes), 1, figsize=(4, 4),dpi=300)

# Plot each attribute
for i, attr in enumerate(attributes):
    axs[i].scatter(df[attr], df['Value'],0.1)
    axs[i].scatter(df_top[attr], df_top['Value'], 2, color='red', label='Top 10 Values')
    axs[i].set_xlabel(attr)
    axs[i].set_ylabel('$\eta$')
    axs[i].set_xlim(ranges[attr])
    #axs[i].set_yscale('log')

plt.tight_layout()
plt.savefig("Grafiken Report/optuna_ox.pdf")
plt.show()
```
